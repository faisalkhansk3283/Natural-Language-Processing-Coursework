```python
import torch

a = torch.tensor(1.0, requires_grad=True)
b = torch.tensor(1.0, requires_grad=True)
x = torch.tensor(-2.0)

f_x = torch.log(a * (x**4) + b * (x**2) - 1)

f_x.backward()

print("df/da:",a.grad)
print("df/db:",b.grad)
print((x**4)/ (a * (x**4) + b * (x**2) - 1 ) )
print((x**2)/ (a * (x**4) + b * (x**2) - 1 ) )
```

```python
import torch

#Given Matrix A of size 2x3
A=torch.tensor([[4.,2.,0.],[-1.,0.,-1.]])

#Matrix A Transpose(AT)
ATran=torch.transpose(A,0,1)
print('Matrix A:\n',A,'\nMatrix A Transpose:\n',ATran)

#A*AT & AT*A
Resultant1_Matrix=torch.matmul(A,ATran)
Resultant2_Matrix=torch.matmul(ATran,A)
print('A * AT:\n',Resultant1_Matrix,'\nAT * A:\n',Resultant2_Matrix)
```
